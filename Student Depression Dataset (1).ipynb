# Student Depression Dataset

#### A student depression dataset typically contains data aimed at analyzing, understanding, and predicting depression levels among students. It may include features such as demographic information (age, gender), academic performance (grades, attendance), lifestyle habits (sleep patterns, exercise, social activities), mental health history, and responses to standardized depression scales.

#### These datasets are valuable for research in psychology, data science, and education to identify factors contributing to student depression and to design early intervention strategies. Ethical considerations like privacy, informed consent, and anonymization of data are crucial in working with such sensitive information.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Elvis\Documents\Student Depression Dataset.csv")
df
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

```python
gender_counts = df['Gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(8, 8)) 
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=0, colors=['skyblue', 'pink']) 
plt.title('Gender Distribution') 
plt.axis('equal')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Dietary Habits', data=df, palette=['red', 'blue', 'black','yellow']) 
plt.title('Gender and Dietary Habit Distribution') 
plt.xlabel('Gender') 
plt.ylabel('Count') 
plt.show()  
```

```python
df.City.unique()
```

```python
gender_counts = df.groupby(['City', 'Gender']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(gender_counts, annot=True, fmt="d", cmap="coolwarm", linewidths=.5)
plt.title('Gender Distribution Across Cities')
plt.xlabel('Gender')
plt.ylabel('City')
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
sns.boxplot(x='Study Satisfaction', y='CGPA', data=df, palette='Set2') 
plt.title('Study Satisfaction vs. CGPA') 
plt.xlabel('Study Satisfaction (1.0: Low, 4.0: High)')
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Financial Stress', y='Work/Study Hours', data=df, palette='Set2')
plt.title('Financial Stress by Work/Study Hours')
plt.xlabel('Financial Stress')
plt.ylabel('Work/Study Hours')
plt.grid(True)
plt.show()
```
